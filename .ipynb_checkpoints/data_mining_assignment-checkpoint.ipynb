```python
# IMPORT PACKAGES 
import pandas as pd;
import numpy as np;
from matplotlib import pyplot as plt;

# REMOVE ALL DEPRECATION WARNINGS
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
# ACCESS THE API URL AND CHECK IF THE RESPONSE IS OK
import requests
import xml.etree.ElementTree as ET  # for parsing XML
import xmltodict
url = "https://www.opensecrets.org/api/?method=getLegislators&id=NJ&&apikey=9c0f1138276886e431b8744bcef794af"
response = requests.get(url)
response
```

```python
root = ET.fromstring(response.content)  # parse XML
```

```python
root.getchildren()
```

```python
# GET DATA COLUMNS TITLES
data_title = root.getchildren()[0].attrib.keys()
data_title
```

```python
# GET DATA
data_list = [];

for child in root.getchildren():
    data_list.append([value for value in child.attrib.values()])
    
legislators_dataframe = pd.DataFrame(data_list,columns=data_title)
legislators_dataframe
```

```python
#STORE THE DATAFRAME DATA TO A POSTGRE DATABASE
import psycopg2

from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:""@localhost:5432/legislators')
legislators_dataframe.to_sql("legislators_table", engine)
```
